```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/mushroom-classification/mushrooms.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
sns.countplot(x = df['class'])
plt.xticks([0,1],['poisonous', 'edible'])
plt.show()
```

```python
plt.figure(figsize = (14,6))
cap_mapping = {
    'b': 'bell', 
    'c': 'conical', 
    'x': 'convex', 
    'f': 'flat', 
    'k': 'knobbed', 
    's': 'sunken'
}

color_map = {
    'n': 'brown', 
    'b': 'buff', 
    'c': 'cinnamon', 
    'g': 'gray', 
    'r': 'green', 
    'p': 'pink', 
    'u': 'purple', 
    'e': 'red', 
    'w': 'white', 
    'y': 'yellow'
}

class_map = {
    'e': 'edible', 
    'p': 'poisonous'
}

df1 = df.copy()
df1['cap-shape'] = df1['cap-shape'].map(cap_mapping)
df1['cap-color'] = df1['cap-color'].map(color_map)
df1['class'] = df1['class'].map(class_map)

g = sns.catplot(
    x='cap-shape',
    hue='cap-color',
    col='class',
    data=df1,
    kind='count',
    height=8,
    aspect=.7
)

g.set_axis_labels("Cap Shape", "Count")
g.set_xticklabels(rotation=45)
g.fig.suptitle("Mushroom Cap Shape and Color Distribution by Class", y=1.02)
plt.tight_layout()
plt.show()
```

1. White colored and Bell Shaped Mushroom are highly recommended for eating
2. Red Coloured Knobbed Shaped Mushroom are poisonous

```python
odor_map = {
    'a' : 'almond',
    'l' : 'anise',
    'c' : 'creosote',
    'y' : 'fishy',
    'f': 'foul',
    'm' : 'musty',
    'n' : 'none',
    'p' : 'pungent',
    's': 'spicy'
}

bruises_map = {
    't': 'bruises',
    'f': 'no'
}

df1['odor'] = df1['odor'].map(odor_map)
df1['bruises'] = df1['bruises'].map(bruises_map)

plt.figure(figsize = (14,6))
g = sns.catplot(
    x = 'odor',
    hue = 'bruises',
    col = 'class',
    data = df1,
    kind = 'count',
    height = 8,
    aspect = .7,
    palette = 'inferno'

)

## odor: almond=a,anise=l,creosote=c,fishy=y,foul=f,musty=m,none=n,pungent=p,spicy=s
## bruises: bruises=t,no=f

```

1. Odourless and Bruised Mushrooms are highly recommended for eating
2. Fishy odour with No Bruises Mushrooms are poisonous

# Data Preprocessing 

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df = df.apply(LabelEncoder().fit_transform)
df.head()
```

```python
x = df.drop(['class'], axis = 1)
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state = 0, max_depth = 5)
dt.fit(x_train, y_train)
```

```python
dt.score(x_train, y_train)
```

```python
predictions = dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

# Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(max_depth = 5)
```

```python
rf.fit(x_train, y_train)
rf.score(x_train, y_train)
```

```python
predictions = rf.predict(x_test)
rf.score(x_test, y_test)
```

```python
accuracy_score(y_test, predictions)
```

1. Decision Tree Accuracy - 98.2
2. Random Forest Accuracy - 99.0

RF is comparatively better!
