# Cleaning up the Data

USA races only, 50k or 50mi, 2020

```python
# 50mi
# 50km
df['Event distance/length'].unique()
```

```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
df2 = df[(df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') & (df['Event distance/length'].isin(['50km', '50mi'])) & (df['Year of event'] == 2020) ]
```

```python
df2.head()
```

Removing US from event name

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

Cleaning Athlete's Age

```python
df2['athlete_age'] = 2025 - df2['Athlete year of birth']
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

Drop Columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category

```python
df2 = df2.drop(columns= ['Athlete club', 'Athlete year of birth','Athlete age category','Athlete country'])
```

```python
df2.head()
```

Clean up null values

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna() == 1]
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

Check for duplicate values

```python
df2[df2.duplicated() == True]
```

Reset Index

```python
df2.reset_index(drop = True)
```

Fix types

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

Rename columns

```python
df2 = df2.rename(columns = {
    'Year of event' :'year',
    'Event dates':'race_day',
    'Event name' :'race_name',
    'Event distance/length' : 'race_length',
    'Event number of finishers':'race_number_of_finishers',
    'Athlete performance' : 'athlete_performance',
    'Athlete gender' : 'athlete_gender',
    'Athlete average speed' :'athlete_average_speed',
    'Athlete ID': 'athlete_id'
})
```

```python
df2.head()
```

Reorder columns 

```python
df3 = df2[['race_day','race_name', 'race_length', 'race_number_of_finishers','athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performance', 'athlete_average_speed']]
```

Difference in speed for the 50k, 50mi, male to female


```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

Age groups that are best in the 50m race (20 + races min) (show 15)

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= False).query('count > 19').head(15)
```

Age groups that are worst in the 50m race (20 + races min) (show 15)

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending= True).query('count > 19').head(15)
```
